**Dictionary keys and values**
* values
  * any type (immutable and mutable)
  * can be duplicates
  * dictionary values can be lists, even other dictionaries!
* keys
  * must be unique
  * immutable type (int, float, string, tuple,bool)
    * actually need an object that is hashable, but think of as immutable as all immutable types are hashable
  * careful with float type as a key
* no order to keys or values!

````
# Example of Fibonacci numbers by recurssion
def fib(x):
  """
  input: a non-negative integer 
  output: the xth Fibonacci number
  """

  if x == 0 or x == 1:
    return 1
  else:
    return fib(x-1)+fib(x-2)


fib(5)
````

````
# Calculate Fibonacci with a dictionary

def fib_efficient(n, d):
  """
  compute the nth Fibonacci number and store it in dictionary d
  """  
  if n in d:
    return d[n] # memorization, creating a memo of what has been done
  else:
    ans = fib_efficient(n-1, d) + fib_efficient(n-2, d)
    d[n] = ans # modify dictioanry as progress through function calls
    return ans

d = {0:1, 1:1} # store the first and second Fibonacci numbers in the dictionary
fib_efficient(5,d)
````
